```python
import pandas as pd
import numpy as np
```

```python
df_2020 = pd.read_csv('players_2020.csv')
```

```python
df_2020.head()
```

```python
grouped_df = df_2020.groupby("federation")
grouped_df.head()
```

```python
sorted_df = grouped_df.sort_values(['federation','rating_standard'], ascending=(True, False))
sorted_df.head(50)
```

```python
drop_df = sorted_df.drop_duplicates(subset='federation', keep='first')
drop_df.head(50)
```

```python
drop_df.to_csv(r'players_by_country.csv', index = False)
```
